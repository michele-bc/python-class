# 6 Propriedades das Funções do Python


1.   Funções atuam como OPERADORES UNÁRIOS de TUPLAS
2.   Funções são OBJETOS e o NOME DE UMA FUNÇÃO é (apenas) uma referência a este objeto
3.   É possível definir uma função DENTRO DE OUTRA FUNÇÃO
4.   Uma função pode ser ARGUMENTO DE OUTRA FUNÇÃO
5.   Uma função pode RETORNAR OUTRA FUNÇÃO
6.    Uma função aninhada TEM ACESSO A OBJETOS da função 'ninho'



## 1. Funções atuam como OPERADORES UNÁRIOS de TUPLAS!!!

```python
# Operador unário: realiza uma operação sobre um único operando
print( - 7 )
print( - 10 )
print()

# Funções: atuam como operadores unários sobre tuplas (os operandos são tuplas)
from math import sqrt, atan2

print( sqrt(4) )
print( sqrt  (4) )
print( sqrt  (4,) )

print( type  ((4)) )
print( type  ((4,)) )    # Type precisa de dois parêntesis e da virgula, pq?
print()

x = 1
y = 1
print( atan2(y, x) )
print( atan2   (y, x) )  # arctan( y/x )
```

## 2. Funções são OBJETOS e o NOME DE UMA FUNÇÃO é (apenas) uma referência a este objeto!!!

```python
# 5 é um objeto armazenado na memória e a é o caminho para o local na memória
a = 5
b = a
print(a, b)
print()


# Funções
from math import sqrt

raiz_quadrada = sqrt

print( sqrt (4), raiz_quadrada (25) )
```

## 3. É possível definir uma função DENTRO DE OUTRA FUNÇÃO!!!

* Funções aninhadas não são acessíveis fora do escopo da função 'ninho'.

```python
# Uma função única:
def soma_um (n):
  resultado = n + 1
  return resultado

print( soma_um (10) )

# Funções aninhadas (função dentro de função):
def soma_um2 (n):                             # Função 'ninho'

  def adiciona_uma_unidade_a_um_numero (x):  # Função aninhada
    return x + 1

  resultado = adiciona_uma_unidade_a_um_numero (n)
  return resultado

print( soma_um2 (5) )

# Funções aninhadas não são acessíveis fora do escopo da função 'ninho'
# print( adiciona_uma_unidade_a_um_numero (2) )  # Erro!
```

## 4. Uma função pode ser ARGUMENTO DE OUTRA FUNÇÃO!!!

```python
def soma_um(n):
  return n + 1

def executa_uma_funcao_f_de_x_qualquer (f, x):
  ''' Executa uma funcao f(x) qualquer. '''
  
  resultado = f(x)
  return resultado

print( executa_uma_funcao_f_de_x_qualquer (soma_um, 10) )

# Exemplo 2:
from math import sqrt

print( executa_uma_funcao_f_de_x_qualquer (sqrt, 25) )
```

## 5. Uma função pode RETORNAR OUTRA FUNÇÃO!!!

* Uma função sempre vai retornar algum objeto, mesmo quando o **return** não está presente, a função retorna o objeto **None**.

```python
def retorna_uma_funcao_que_diz_Ni ():

  def diz_Ni ():
    print('Ni!')
    return ''

  return diz_Ni


f = retorna_uma_funcao_que_diz_Ni ()  # Outro identificador para a função
print( type(f) )
print( type(f ()) )
print()
print( f () )                         # O que ela faz e o que retorna
print('.')
```

## 6. Uma função aninhada TEM ACESSO A OBJETOS da função 'ninho'!!!

```python
def exibe_mensagem (texto):

  def imprime_mensagem ():
    print(' Mensagem: ' + texto)

  imprime_mensagem ()


exibe_mensagem('Funcionou!!')
```
