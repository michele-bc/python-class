# Iteradores (iterators)

* Objetos associados a coleções de dados utilizados em iterações de elementos de coleções de dados.



## Iterações com "for"

```python
string = 'abcdef'
for c in string:
  print(c)
print()

conjunto = {0, 10, 3, 4, 5}
for x in conjunto:
  print(x)
```

## Iterações com "while"

```python
string = 'abcdef'
k = 0
while k < len(string):
  print( string[k] )
  k += 1
```

## Método especial: \_\_iter\_\_()

* O método \_\_iter\_\_ está definido apenas para objetos iteráveis (*strings*, listas, conjuntos, ...), retornando exceção para objetos atômicos.

```python
string = 'abcdef'

it = string.__iter__()

print('  string =', string)
print('      it =', it)
print('type(it) =', type(it))

print()

lista = [0, 1, 2, 3, 4]
itx = lista.__iter__()

print('    lista =', lista)
print('      itx =', itx)
print('type(itx) =', type(itx))
```

```python
# __iter__ não está definido para números inteiros pq eles não são iteráveis
# e retorna uma exceção
x = 10
print('type(x) =', type(x))

itint = x.__iter__()  # Erro!
```

## Método especial (do iterador): \_\_next\_\_()

* Objetos de coleções de dados retornam o iterador através do método especial **\_\_iter\_\_**, que contém uma cópia da coleção de dados e um contador que será utilizado nas iterações.

* Ele retorna cada elemento da coleção de dados na sequência através do método especial do iterador **\_\_next\_\_**.

```python
string = 'abcdef'       # Objeto contendo uma coleção de dados

it = string.__iter__()  # Obtendo o iterador do objeto 'string'

print( it.__next__() )  # Retorna o 1o objeto da coleção de dados

print( it.__next__() )  # Retorna o elemento seguinte da coleção de dados
print( it.__next__() )  # Retorna o elemento seguinte da coleção de dados
print( it.__next__() )  # Retorna o elemento seguinte da coleção de dados
print( it.__next__() )  # Retorna o elemento seguinte da coleção de dados

print( it.__next__() )  # Retorna o 6o e último elemento da coleção de dados

print( it.__next__() )  # Erro! Tentando obter um elemento além do último
```

## Exemplo de uso dos métodos especiais de iteradores

```python
string = 'abcdef'

it = string.__iter__()

while True:
  try:
    c = it.__next__()
    print(c)
  except StopIteration:
    break
```

```python
# O for vai fazer uma cópia da coleção de dados, escrever um while utilizando
# uma estrutura try, obtendo o elemento com o método __next__() e quando o
# iterador retorna uma exceção, ele interrompe a iteração, ou seja, o for usa
# uma estrutura semelhante à anterior
string = 'abcdef'

for c in string:
  print(c)
```

## Funções: iter() e next()

* Não é boa prática de programação utilizar os métodos **\_\_iter\_\_** e **\_\_next\_\_**. No lugar, pode-se usar as funções **iter()** e **next()** que chamam esses métodos.

* Não existe uma função ou método, por exemplo, **previous()**. As iterações são sempre no sentido progressivo e nunca no sentido reverso. Para iterar no sentido reverso, usa-se a função **reversed()** ao criar o **iter**.

* Não usar iteradores em objetos iteráveis porque o **for** já faz isso. O uso é para objetos que não são iteráveis ou dentro da definição de novas classes para que elas sejam iteráveis pelo **for**.

```python
string = 'abcdef'

it = iter(string)        # iter(string) ==> string.__iter__()

while True:
  try:
    c = next( it )
    print(c)
  except StopIteration:  # next(string) ==> string.__next__()
    break
```

```python
# Iteração no sentido reverso
string = 'abcdef'

it = iter( reversed(string) )
while True:
  try:
    c = next( it )
    print(c)
  except StopIteration:
    break
```

## Exemplo: Gás de Partículas

* Perguntas básicas:

  * O que, EXATAMENTE, o código deve fazer?

    > O código deve modelar um sistema (gás) formado por n partículas, sendo que cada partícula possui uma massa m, está em uma posição (x,y) e tem uma velocidade v = (vx,vy).

  * De que DADOS o código precisa para fazer o que deve fazer?

    * Número de partículas, n
    * Distribuição de massas (iguais?, aleatórias?)
    * Distribuição das posições (uniforme?, aleatórias?)
    * Distribuição das velocidades (aleatórias?, M-B?)

  * O que, EXATAMENTE, o código deve produzir como resultado (*output*)?

    * Gráfico mostrando as posições das partículas

  * Qual a abordagem a ser usada no desenvolvimento do código: programação estruturada ou programação orientada a objetos?

    * POO (programação orienta a objetos)

```python
# O nome de classes devem começar com letras maiúsculas
class Particle:
  ''' A particle of the gas. '''

  slots = ['m', 'x', 'y', 'vx', 'vy']

  def __init__(self, m, x, y, vx, vy):
    ''' Initialize the attributes. '''
    self.m = m
    self.x = x
    self.y = y
    self.vx = vx
    self.vy = vy

  def __repr__(self):
    return 'Particle: (  m = ' + str(self.m).ljust(22) + \
                       ' x = ' + str(self.x).ljust(22) + \
                       ' y = ' + str(self.y).ljust(22) + \
                       'vx = ' + str(self.vx).ljust(22) + \
                       'vy = ' + str(self.vy).ljust(22) + ')'



class GasOfParticles:
  ''' Model a gas with n particles. '''

  def __init__(self, n=0):
    ''' Initialize the attributes of the object. '''
    self.n = n
    self.listp = list()
    self.__createParticles()

  def __iter__(self):
    ''' Returns an iterator. '''
    self.k = 0
    return self

  def __next__(self):
    ''' Returns the current particle. '''
    if self.k < len(self.listp):
      p = self.listp[self.k]
      self.k += 1
      return p
    else:
      raise StopIteration

  def __createParticles(self):
    ''' Create a list of particles. '''
    from random import random
    for i in range(self.n):
      m = random()
      x = random()
      y = random()
      vx = -1 + 2 * random()
      vy = -1 + 2 * random()
      p = Particle(m, x, y, vx, vy)
      self.listp.append(p)

  def __getCoordinate(self, coord='x'):
    ''' Returns a list with x or y. '''
    lista = list()
    for p in self.listp:
      if coord == 'x':
        lista.append( p.x )
      else:
        lista.append( p.y )
    return lista

  def plot(self):
    ''' Build a plot with the particle distribution. '''
    import matplotlib.pyplot as plt
    plt.figure(dpi=100)
    listx = self.__getCoordinate('x')
    listy = self.__getCoordinate('y')
    plt.plot(listx, listy, 'or')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Particle distribution")
    plt.show()



# Main program:
gas = GasOfParticles(n=10)
gas.plot()

for p in gas:
  print(p)
```
