## Gerando arranjos numéricos

* **np.arange(x_i, x_f, dx)**: cria um arranjo com valores de x_i até antes de x_f com incremento que pode ser fracionário dx.

* **np.linspace(x_i, x_f, n)**: cria um arranjo com valores de x_i até (igual a) x_f dividindo o intervalo igualmente em n pontos.

* **np.zeros(n)**: cria um arranjo com n zeros.

* **np.ones(n)**: cria um arranjo com n elementos iguais a 1.

* **np.random.randint(x_i, x_f, n)**: retorna um arranjo com n números pseudo-aleatórios inteiros de x_i até antes de x_f.

* **np.random.rand(n)**: retorna um arranjo com n números pseudo-aleatórios reais de 0 até antes de 1.

```python
import numpy as np

v = np.array( [1,2,3,4], dtype='float64' )
print("v =", v)
```

```python
# range():
a = np.array( [x for x in range(10)] )
print('a =', a, '\n')

a = np.array( range(-20, 30, 3) )
print('a =', a, '\n')

r = range(-20, 30, 3)
print(r, type(r))                               # range() não é uma lista
print()

# np.arange(x_i, x_f, dx):
print( "np.arange() =", np.arange(0, 5, 0.2) )  # Diferentemente de range(), np.arange() trabalha com steps reais
print()

# np.linspace(x_i, x_f, n):
print( "np.linspace() =", np.linspace(0, 5, 50) )
```

```python
# np.zeros(n):
a = np.zeros(10)
print('a =', a)
```

```python
# np.ones(n):
a = np.ones(10)
print('a =', a)
```

```python
# Números aleatórios:
# from numpy import random

a = np.random.randint(1, 100, 10)
print('a =', a, '\n')

a = np.random.rand(10)
print('a =', a, '\n')
```

## Iterando elementos de um array nD

* **arranjo.flat**: retorna um objeto iterável com os elementos do arranjo.

* **np.nditer(arranjo)**: retorna um objeto iterável com os elementos do arranjo. 

* **np.ndenumerate(arranjo)**: retorna uma arranjo contendo tuplas dos índices e elementos de um arranjo.

```python
a = np.arange(1, 21).reshape(4, 5)
print('a =\n', a, '\n')

# 1a. tentativa:
for elemento in a:
  print(" Elemento:", elemento)
print()

# Transformando tensores em vetores para poder iterar diretamente:
for elemento in a.flat:
  print(" Elemento:", elemento)
print()

print( a.flat, type(a.flat) )
```

```python
# np.nditer():
print(np.nditer(x), type(np.nditer(x)))
for x in np.nditer(a):
  print(x)
print()

# np.ndenumerate():
for i, x in np.ndenumerate(a):
  print(x)
```

## Operações com arranjos

* Binárias:
  * **np.add(arranjo1, arranjo2)**: retorna um arranjo com a soma elementos de mesmo índice de dois arranjos que possuam o mesmo *shape*. Equivalente: **arranjo1 + arranjo2**.
  * **np.subtract(arranjo1, arranjo2)**: retorna um arranjo com a subtração dos elementos de mesmo índice de dois arranjos que possuam o mesmo *shape*. Equivalente: **arranjo1 - arranjo2**.
  * **np.divide(arranjo1, arranjo2)**: retorna um arranjo com a divisão dos elementos de mesmo índice de dois arranjos que possuam o mesmo *shape*. Equivalente: **arranjo1 / arranjo2**.
  * **np.multiply(arranjo1, arranjo2)**: retorna um arranjo com a multiplicação dos elementos de mesmo índice de dois arranjos que possuam o mesmo *shape*. Equivalente: **arranjo1 * arranjo2**.
  * **np.dot(arranjo1, arranjo2)**: retorna um arranjo com o produto interno de duas matrizes. Equivalente: **arranjo1 @ arranjo2** e **arranjo1.dot(arranjo2)**.


* Unárias:
  * **np.transpose(arranjo)**: retorna a matriz transposta do arranjo. Equivalente: **arranjo.T**.
  * **np.sum(arranjo)**: retorna a soma de todos os elementos de um arranjo.
  * **np.sum(arranjo, axis=)**: retorna a soma de todos os elementos das linhas (**axis=1**) ou das colunas (**axis=0**) de um arranjo.
  * **np.mean(arranjo)**: retorna a média de todos os elementos de um arranjo.
  * **np.mean(arranjo, axis=)**: retorna a média de todos os elementos das linhas (**axis=1**) ou das colunas (**axis=0**) de um arranjo.
  * **np.max(arranjo)**: retorna o maior valor de todos os elementos de um arranjo.
  * **np.max(arranjo, axis=)**: retorna o maior valor de todos os elementos das linhas (**axis=1**) ou das colunas (**axis=0**) de um arranjo.
  * **np.min(arranjo)**: retorna o menor valor de todos os elementos de um arranjo.
  * **np.min(arranjo, axis=)**: retorna o menor valor de todos os elementos das linhas (**axis=1**) ou das colunas (**axis=0**) de um arranjo.

* Com escalares:
  * **número * arranjo** ou **arranjo * número**
  * **número / arranjo** ou **arranjo / número**
  * **número + arranjo** ou **arranjo + número**
  * **número - arranjo** ou **arranjo - número**

### Operações binárias

```python
# Soma de duas matrizes:
x = np.array( [1,2,3,4] ).reshape(2,2)
y = np.array( [7,8,9,10] ).reshape(2,2)

print('x =\n', x, '\n')
print('y =\n', y, '\n')

z = np.add(x, y )  # Precisam ter o mesmo shape
print('z =\n', z, '\n')

w = x + y
print('w =\n', w, '\n')
```

```python
# Subtração de duas matrizes:
x = np.array( [1,2,3,4] ).reshape(2,2)
y = np.array( [7,8,9,10] ).reshape(2,2)

print('x =\n', x, '\n')
print('y =\n', y, '\n')

z = np.subtract(x, y)  # Precisam ter o mesmo shape
print('z =\n', z, '\n')

w = x - y
print('w =\n', w, '\n')
```

```python
# Subtração de duas matrizes:
x = np.array( [1,2,3,4] ).reshape(2,2)
y = np.array( [7,8,9,10] ).reshape(2,2)

print('x =\n', x, '\n')
print('y =\n', y, '\n')

z = np.subtract(y, x)  # Precisam ter o mesmo shape
print('z =\n', z, '\n')

w = y - x
print('w =\n', w, '\n')
```

```python
# Divisão de duas matrizes:
x = np.array( [1,2,3,4] ).reshape(2,2)
y = np.array( [7,8,9,10] ).reshape(2,2)

print('x =\n', x, '\n')
print('y =\n', y, '\n')

z = np.divide(x, y)  # Precisam ter o mesmo shape
print('z =\n', z, '\n')

w = x / y
print('w =\n', w, '\n')
```

```python
# Multiplicação de duas matrizes:
x = np.array( [1,2,3,4] ).reshape(2,2)
y = np.array( [7,8,9,10] ).reshape(2,2)

print('x =\n', x, '\n')
print('y =\n', y, '\n')

z = np.multiply(x, y)  # Precisam ter o mesmo shape
print('z =\n', z, '\n')

w = x * y
print('w =\n', w, '\n')
```

```python
# Produto interno de matrizes:
x = np.array( [1,2,3,4] ).reshape(2,2)
y = np.array( [7,8,9,10] ).reshape(2,2)

print('x =\n', x, '\n')
print('y =\n', y, '\n')

z = np.dot(x, y)
print('z =\n', z, '\n')

z = x.dot(y)
print('z =\n', z, '\n')

w = x @ y
print('w =\n', w, '\n')
```

### Operações unárias

```python
# Transposta de uma matriz:
x = np.array( [1,2,3,4] ).reshape(2,2)

print('x =\n', x, '\n')

print("Transposta(x) =\n", np.transpose(x), '\n')

print("Transposta(x) =\n", x.T, '\n')

u = np.array( [1,2,3] ).reshape(1,3)

print('x =\n', u, '\n')

print("Transposta(x) =\n", np.transpose(u), '\n')
```

```python
# Soma de todos os elementos de uma matriz:
x = np.array( [1,2,3,4] ).reshape(2,2)

print('x =\n', x, '\n')

print('Soma(x) =', np.sum(x), '\n')
print('Soma(x) =', np.sum(x, axis=0), '\n')  # Soma elementos das linhas
print('Soma(x) =', np.sum(x, axis=1), '\n')  # Soma elementos das colunas

# Média de todos os elementos da matriz:
print('Média(x) =', np.mean(x), '\n')
print('Média(x) =', np.mean(x, axis=0), '\n')  # Soma elementos das linhas
print('Média(x) =', np.mean(x, axis=1), '\n')  # Soma elementos das colunas

# Maior valor de todos os elementos da matriz:
print('Maior(x) =', np.max(x), '\n')
print('Maior(x) =', np.max(x, axis=0), '\n')  # Soma elementos das linhas
print('Maior(x) =', np.max(x, axis=1), '\n')  # Soma elementos das colunas

# Maior valor de todos os elementos da matriz:
print('Menor(x) =', np.min(x), '\n')
print('Menor(x) =', np.min(x, axis=0), '\n')  # Soma elementos das linhas
print('Menor(x) =', np.min(x, axis=1), '\n')  # Soma elementos das colunas
```

## Sistemas de equações lineares

$$\alpha + 3\, \beta = 1$$
$$2\,\alpha - \beta = -2$$

Deseja-se encontrar os valores de $\alpha$ e $\beta$.

Reescrever o sistema de equações na forma matricial:

$$
\begin{bmatrix}
1 & 3 \\
2 & -1
\end{bmatrix}
\cdot
\begin{bmatrix}
\alpha \\
\beta
\end{bmatrix}
=
\begin{bmatrix}
1 \\
-2
\end{bmatrix}
$$

ou 

$${\bf A} \cdot {\bf x} = {\bf b} $$

onde

$$
{\bf A} \equiv
\begin{bmatrix}
1 & 3 \\
2 & -1
\end{bmatrix}
$$

$$
{\bf x} \equiv
\begin{bmatrix}
\alpha \\
\beta
\end{bmatrix}
$$

$$
{\bf b} \equiv
\begin{bmatrix}
1 \\
-2
\end{bmatrix}
$$

* Para sistemas de equações lineares é preciso verificar se o determinante é diferente de zero para ter soluções. Se for igual a zero o sistema não tem solução.

```python
# Implementando via NumPy:
A = np.array( [1,3,2,-1] ).reshape(2,2)
b = np.array( [1,-2] ).reshape(2,1)

print("A =\n", A, '\n')
print("b =\n", b, '\n')

detA = np.linalg.det(A)

print('det(A) =', detA, '\n')  # É diferente de zero então possui solução

x = np.linalg.solve(A, b)

alpha = x[0][0]
beta = x[1][0]
print('x =\n', x)
print('\talpha =', alpha)
print('\t beta =', beta, '\n')

# Teste 1:
print( 'alpha + 3 * beta =', alpha + 3 * beta)
print( '2 * alpha - beta =', 2 * alpha - beta, '\n' )

# Teste 2:
print( 'A @ x =\n', A @ x )
```

## Matriz Inversa

$$ {\bf A} \cdot {\bf x} = {\bf b} $$

$$ {\bf A^{-1}} \cdot {\bf A} \cdot {\bf x} = {\bf A^{-1}} \cdot {\bf b} $$

mas ${\bf A^{-1}} \cdot {\bf A} = {\bf I}$, então

$$ {\bf I}\cdot {\bf x} = {\bf A^{-1}} \cdot {\bf b} $$

$$ {\bf x} = {\bf A^{-1}} \cdot {\bf b} $$

```python
# Implementando via NumPy:
A = np.array( [1,3,2,-1] ).reshape(2,2)
b = np.array( [1,-2] ).reshape(2,1)

Ainv = np.linalg.inv(A)

print('Ainv =\n', Ainv, '\n')

# Teste:
print( Ainv @ A, '\n' )

x = Ainv @ b

alpha = x[0][0]
beta = x[1][0]
print('x =\n', x)
print('\talpha =', alpha)
print('\t beta =', beta, '\n')

# Teste 1:
print( 'alpha + 3 * beta =', alpha + 3 * beta)
print( '2 * alpha - beta =', 2 * alpha - beta, '\n' )

# Teste 2:
print( 'A @ x =\n', A @ x )
```
