# Controle de fluxo

O Python executa cada linha de código sequencialmente, mas as vezes é interessante que algumas partes do código sejam executadas apenas se uma condição for satisfeita ou repetidamente até que uma condição seja satisfeita.

* Ramificação condicional: **if**
* *Loop*: **while**, **for**


## If

Ramificação condicional. Uma suíte¹ será executada caso uma condição seja satisfeita. 

¹ Suíte: conjunto de uma ou mais linhas de código
   código dentro de uma cláusula.

* Sintaxe:

```
if( condição ):  # Cláusula
   suíte a ser executada quando a condição é verdadeira
```

```
if( condição ):  # Cláusula
   suíte a ser executada quando a condição é verdadeira
else:
  suíte a ser executada quando a condição é falsa
```

```
if( condição 1 ):  # Cláusula
   suíte a ser executada quando a condição 1 é verdadeira
elif( condição 2 ):
  suíte a ser executada quando a condição 1 é falsa, mas a 2 é verdadeira
else:
  suíte a ser executada quando ambas as condições são falsas
```


Obs: atenção à identação! Ela é obrigatória para a suíte dentro do if e precisa seguir uma única identação: se usar espaços, precisa sempre usar espaços e se usar tab, só pode usar tab.

````
# if:
# 1 condição e 1 caminho possível

x = 5 #@param

if( x > 0 ):   # Cláusula 1: x>0
  y = x
  print(x, "é positivo e seu valor absoluto é:", y)
````

````
# if-else:
# 1 condição e 2 caminhos possíveis

x =  -1#@param

if( x > 0 ):   # Cláusula 1: x>0
  y = x
  print(x, "é positivo e seu valor absoluto é:", y)
else:
  y = -x
  print(x, "é negativo e seu valor absoluto é:", y)
````

````
# if-elif:
# Qual é o quadrante de (x,y)?

x = -1 #@param
y = -1 #@param

if(x == 0 and y == 0):
  quadrante = 0
elif(x >= 0 and y > 0):
  quadrante = 1
elif(x < 0 and y >= 0):
  quadrante = 2
elif(x < 0 and y < 0):
  quadrante = 3
elif(x >= 0 and y < 0):  # É desnecessária, pode ser substituída por else
  quadrante = 4

print(x, y, "quadrante:", quadrante)
````

````
# if-elif:
# Qual é o quadrante de (x,y)?

# O else captura tudo o que não foi capturado pelas condições anteriores
# Portanto, como só existem as 5 alternativas e 4 foram testadas anteriormente,
# a que restou é a verdadeira
 
x = -1 #@param
y = -1 #@param

if(x == 0 and y == 0):
  quadrante = 0
elif(x >= 0 and y > 0):
  quadrante = 1
elif(x < 0 and y >= 0):
  quadrante = 2
elif(x < 0 and y < 0):
  quadrante = 3
else:
  quadrante = 4

print(x, y, "quadrante:", quadrante)
````

````
# Aninhamento de ifs:
 
x = -1 #@param
y = -1 #@param

if(x > 0):
  if(y > 0):
    quadrante = 1
  else:
    quadrante = 4
else:
  if(y > 0):
    quadrante = 2
  else:
    quadrante = 3

print(x, y, "quadrante:", quadrante)
````

````
# if em expressões condicionais:
# Qual é o quadrante em que (x,y) está?
# Quadrante é 0 se x e y estão na origem senão
#             1 se x > 0 e y > 0
#             2 se x > 0 e y > 0
#             3 se x < 0 e y < 0
#             4 se x > 0 e y < 0

x = -1 #@param
y = -1 #@param

if(x > 0):
  if(y > 0):
    quadrante = 1
  else:
    quadrante = 4
elif(x < 0):
  if(y > 0):
    quadrante = 2
  else:
    quadrante = 3
else:
  quadrante = 0

print(x, y, "quadrante:", quadrante)
````

````
# if em expressões condicionais:
# Qual é o quadrante em que (x,y) está?
# Quadrante é 0 se x e y estão na origem senão
#             1 se x > 0 e y > 0
#             2 se x > 0 e y > 0
#             3 se x < 0 e y < 0
#             4 se x > 0 e y < 0

x =  4#@param
y =  -2#@param

quadrante = 0 if x == 0 and y == 0 else \
            1 if x >= 0 and y > 0 else \
            2 if x < 0 and y >= 0 else \
            3 if x < 0 and y < 0 else \
            4

print(x, y, "quadrante:", quadrante)
````

````
# Par ou ímpar?
# tipo é "par" se x for divisível por zero, senão "ímpar"

x = 4 #@param

tipo = "par" if (x % 2 == 0) else "ímpar"

print(x, tipo)
````

````
bissexto = True #@param {type:"boolean"}

numero_de_dias_do_mes = [0, 31, 29 if bissexto else 28,
                         31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

print(numero_de_dias_do_mes)
````

## While

Enquanto uma certa condição for válida, execute a suíte.

* Sintaxe:

```
while( condição ):
  suíte a ser executada caso a condição seja verdadeira
```

```
while( condição ):
  suíte a ser executada caso a condição seja verdadeira
else:
  suíte a ser executada caso a condição seja falsa
```

Também é possível fazer *loops* infinitos/indefinidos.

* Sintaxe:

```
while True:
  if( condição ):
    break     # Quebra o loop
  else:
    continue  # Não faz nada e continua o loop
```

```
condicao_loop = True
while condicao_loop:
  suíte
  if( condição ):
    condicao_loop = False
```




````
# while:

k = 0

while k <= 10:
  print(k)
  k += 1
````

````
# while-else:

k = 20

while k <= 10:
  print(k)
  k += 1
else:
  print("o loop while não foi executado")

print("executa o resto do código")
````

````
# loop infinito

print('\n',
      " [a] ... faz uma coisa...", '\n',
      " [b] ... faz uma outra coisa... \n",
      " [c] ... faz uma terceira coisa...")

opcao = input("Digite sua opção: ")

if opcao == "a":
  print(" Fiz uma coisa...")
elif opcao == 'b':
  print(" Fiz a outra coisa...")
elif opcao == 'c':
  print(" Fiz uma terceira coisa...")
else:
  print(" O usuário digitou um valor inválido.")
````

````
# loop infinito

print('\n',
      " [a] ... faz uma coisa...", '\n',
      " [b] ... faz uma outra coisa... \n",
      " [c] ... faz uma terceira coisa...")

while True:
  opcao = input("Digite sua opção: ")
  if( opcao == 'a' or opcao == 'b' or opcao == 'c' ):
    break     # Quebra o loop
  else:                                     # <--- pq usar se sem ele funciona?
    continue  # Não faz nada e continua o loop

if opcao == "a":
  print(" Fiz uma coisa...")
elif opcao == 'b':
  print(" Fiz a outra coisa...")
elif opcao == 'c':
  print(" Fiz uma terceira coisa...")
else:
  print(" O usuário digitou um valor inválido.")
````

````
# loop indefinido

print('\n',
      " [a] ... faz uma coisa...", '\n',
      " [b] ... faz uma outra coisa... \n",
      " [c] ... faz uma terceira coisa...")

valor_invalido = True
while valor_invalido:
  opcao = input("Digite sua opção: ")
  if( opcao == 'a' or opcao == 'b' or opcao == 'c' ):
    valor_invalido = False

if opcao == "a":
  print(" Fiz uma coisa...")
elif opcao == 'b':
  print(" Fiz a outra coisa...")
elif opcao == 'c':
  print(" Fiz uma terceira coisa...")
else:
  print(" O usuário digitou um valor inválido.")
````

## For

É um comando de *loop* que repete uma suíte e precisa de um objeto iterado, por exemplo, uma lista.

O **for** serve para trabalhar com cada um dos elementos de objetos iterados (p.e. listas).

* Sintaxe:

```
for elemento in coleção:
  suíte
```

```
for elemento in coleção:
  suíte
else:
  suíte
```

O objeto iterado armazena a informação do último objeto que foi iterado e pega o próximo até que chegue ao último elemento e a lista informa ao **for** que todos os elementos já foram liberados.

Pode-se usar o **for** com a função **range()** (que é uma classe como int, str, etc) principalmente para determinar um incremento ao selecionar os elementos.

A função **range()** retorna uma sequência de números, começando pelo valor mínimo (ou 0 por padrão) até antes do valor máximo com um determinado incremento (1 por padrão).

* Sintaxe:
```
range(valor mínimo, valor máximo, incremento)
```

* Sintaxe do uso no **for**:
```
for i in range(v_min,v_max,step):
  suíte
```

````
# for em lista

lista = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', "Jun"]

for mes in lista:
  print(mes)
````

````
# for  em lista
# Também é possível escrever a lista, não é necessário que seja uma variável

for mes in ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', "Jun"]:
  print(mes)
````

````
# for em lista
# Outra forma possível é utilizando range(v_min,v_max,step) que retorna
# os valores de v_min a (v_min - 1) que incrementam em step (step default = 1).

lista = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', "Jun"]

for i in range(0, 6):
  mes = lista[i]
  print(mes)
````

````
# for em strings

alfabeto = 'ABCDEFGHIJKLMNOPQRSTUVXWYZ'

for letra in alfabeto:
  print(letra)

print()

# Também é possível usar o range para escolher elementos que variam com 
# determinado incremento na posição
n = len(alfabeto)

for i in range(0, n, 4):
  print( alfabeto[i] )
````

````
# for-else

lista = list()

for elemento in lista:
  print(" Elemento: *", elemento, "*")
else:
  print(" Nenhuma iteração foi realizada.")
````
