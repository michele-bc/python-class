# Módulos

* Contêm pedaços de códigos que podem ser reutilizados por diferentes programas.

* Para possibilitar que uma função, por exemplo, possa ser chamada por qualquer programa, sua definição pode ser salva em um arquivo (*.py) junto com a definição de outras funções afins.

* Várias definições de funções em um arquivo constiuem um "módulo" que pode ser importado por qualquer programa.

* O módulo "math" são arquivos (*.py) contendo definições de funções Python.

* Vários módulos podem ser agrupados dentro de uma "biblioteca".

* Por exemplo, a biblioteca "matplotlib" contém, entre outros, o módulo "pyplot".

* Em nivel de Sistema Operacional, uma biblioteca Python é um diretório contendo arquivos *.py (módulos) e um arquivo especial que indica ao Python que aquele diretório é uma biblioteca.






## Módulos nativos (built-in modules)

* Módulos que acompanham toda distribuição Python.

* **help(módulo)**: função para obter informações sobre módulos.

````python
import math
help(math)
````

## Importando um módulo

* Sintaxe:

```
import módulo
```

* É boa prática de programação colocar as importações de módulos no início do código, mas elas podem ser colocadas em qualquer lugar desde que antes do uso de suas definições.

````python
# Importando um módulo por vez (através do comando import)
import math   # Funções matemáticas
import cmath  # Funções matemáticas complexas
import sys    # Funções que atuam sobre o sistema
````

````python
# Importando um mesmo módulo mais de uma vez
import math
import math
````

## Acessando objetos contidos em um módulo

* Sintaxe: qualificação.definição

```
módulo.função()
módulo.constante
```

````python
# sqrt() : uma das funções do módulo math
# math.  : qualificação da função sqrt(), o módulo a quem a função pertence

print( math.sqrt(9) )
````

## Importando vários módulos de uma única vez

* Sintaxe:

```
import módulo1, módulo2, ...
```

````python
import math, cmath, sys
````

## Importando um módulo e alterando o seu nome

* Sintaxe:

```
import módulo as novo_nome
```

* É útil quando um módulo é muito usado ao longo do programa ou o nome do módulo é muito grande e um nome menor facilita a escrita do programa.

````python
import math as m

print( m.sqrt(16) )
print( m.cos( m.pi ) )
````

````python
print( math.sqrt(25) )  # Funciona porque o módulo 'math' foi importado anteriormente
````

## Deletar módulos importados

* Sintaxe:

```
del módulo
```

````python
del math
del cmath
del sys
````

````python
import math
print( math.sqrt(25), m.sqrt(25) )  # Não é boa prática de programação manter módulo com dois nomes
````

````python
import matplotlib.pyplot as plt
# matplotlib : biblioteca (de módulos)
# pyplot     : módulo (com funções gráficas)

x = [-3, -2, -1, 0, 1, 2, 3]
y = [9, 4, 1, 0, 1, 4, 9]

plt.plot(x, y)
````

## Importando um objeto específico de um módulo

* Sintaxe: 

```
from módulo import objeto
```

* Nesse tipo de importação não é preciso qualificar o objeto.

* Não é recomendado porque os nomes dos objetos podem ser usados ao longo do código para referenciar outra coisa e não mais o objeto.

````python
from math import sqrt

print( sqrt(16) )
````

````python
from math import sqrt, cos, sin

print( cos(3.14), sin(3.14), sqrt(15) )
````

````python
from math import (sqrt, cos, sin)

print( cos(3.14), sin(3.14), sqrt(15) )
````

````python
# PERIGO !
sqrt = 'abc'
cos = 'def'
sin = 'ab'

print(sqrt, cos, sin)
````

## Importando todos os objetos de um mesmo módulo

* Sintaxe:

```
from módulo import *
```

* Não é preciso qualificar o objeto também. Cuidado ao usar.

````python
from math import *
````

````python
print( cos(-1), acos(-1), sqrt(18), atan(0.5) )
````

## PYTHONPATH

* Ao importar um ḿodulo, por exemplo o math, o Python procura pelo arquivo 'math.py' no diretório onde está o programa.

* Se ele encontrar dentro do diretório, o Python importa todos objetos definidos.

* Se ele não encontra, ele procura nos diretórios de uma variável ambiente do linux chamada PYTHONPATH, que é uma lista de diretórios onde estão módulos do Python.

````python
# O '!' no começo da linha indica que iremos acessar o linux
! ls
! pwd
print()
! printenv PYTHONPATH  # É onde o Python procura os módulos: /env/python
````

## Diretórios do sistema

* Se ele não encontrar dentro do PYTHONPATH ele procura dentro dos diretórios do sistema.

````python
import sys

print( sys.path )  # É onde o Python procura os módulos
````

````python
! ls /usr/lib/python3/dist-packages
````

````python
# Retorna exceção se tenta importar um módulo que não existe
import xyz
````

* Existe uma hierarquia ao procurar módulos:
  * Diretório onde está o programa
  * PAYTHONPATH
  * Diretórios do sistema

* É preciso tomar cuidado ao definir um módulo para que ele não contenha o mesmo nome de módulos nativos, pois se tiver, não poderá usar os módulos nativos.

* Se não existe o módulo importado, o Python retorna uma exceção.
