## Gráficos xy 

```python
import matplotlib.pyplot as plt

# Gerando dados:
x = [10,20,30,40]
y = [100,400,900,1600]

# plot(x, y):
plt.plot(x, y)  # O número de elementos das duas listas precisam ser iguais senão gera exceção

plt.title('Gráfico 1')
plt.xlabel('x')
plt.ylabel("y")

plt.show()

# plot(y):
plt.plot(y, 'o-')  # Quando x é omitido o Python usa os índices da lista

plt.title('Gráfico 1b')
plt.xlabel('x')
plt.ylabel("y")

plt.show()
```

### Múltiplas curvas num mesmo gráfico

* Uma forma é plotar em linhas diferentes e no último usar **plt.show()**.

* Outra forma é plotar na mesma linha **plt.plot(x1,y1, x2,y2, x3,y3, ...)**

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando os dados:
x = np.arange(-np.pi, np.pi, 0.1)
y1 = x
y2 = np.sin(x)
y3 = np.cos(x)

# Modo 1:
plt.plot(x, y1)            # Azul
plt.plot(x, y2)            # Laranja
plt.plot(x, y3)            # Verde
plt.plot(x, np.cos(x)**2)  # Vermelha

plt.title("Gráfico 2a")
plt.xlabel('x')
plt.ylabel("y")
plt.show()

# Modo 2:
plt.plot(x,y1, x,y2, x,y3, x,np.cos(x)**2)

plt.title("Gráfico 2b")
plt.xlabel('x')
plt.ylabel("y")
plt.show()
```

```python
# Modo 2 fundiona também para funções com diferentes abcissas:
x1 = range(10)
y1 = [ x**2 for x in x1 ]
x2 = range(8)
y2 = [ -x for x in x2 ]
x3 = range(6)
y3 = [ x+2 for x in x3 ]

plt.plot(x1,y1, x2,y2, x3,y3)

plt.title("Gráfico 2b")
plt.xlabel('x')
plt.ylabel("y")
plt.show()
```

### Alterando a espessura das curvas

* Usa argumento da função **plt.plot()** lw = line width (*default* é 1).

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando os dados:
x = np.arange(-np.pi, np.pi, 0.1)
y1 = x
y2 = np.sin(x)
y3 = np.cos(x)

# Modo 1:
# lw = nume width (lw=1, por default)
plt.plot(x, y1, lw=1)            # Azul
plt.plot(x, y2, lw=2)            # Laranja
plt.plot(x, y3, lw=3)            # Verde
plt.plot(x, np.cos(x)**2, lw=4)  # Vermelha

plt.title("Gráfico 3a")
plt.xlabel('x')
plt.ylabel("y")
plt.show()

# Modo 2:
# Nesse modo não é possível alterar a largura de cada linha individualmente
plt.plot(x,y1, x,y2, x,y3, x,np.cos(x)**2, lw=4)

plt.title("Gráfico 3b")
plt.xlabel('x')
plt.ylabel("y")
plt.show()
```

### Alterando o estilo das curvas

* Cores:
  * r = red
  * g = green
  * b = blue
  * y = yellow
  * c = cyan
  * m = magenta
  * k = black
  * w = white

* Estilo:
  * Sólida      = - (*default*)
  * Tracejada   = --
  * Traço-ponto = -.
  * Pontilhada  = :

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando os dados:
x = np.arange(-np.pi, np.pi, 0.1)
y1 = x
y2 = np.sin(x)
y3 = np.cos(x)

# Modo 1:
# lw = nume width (lw=1, por default)
plt.plot(x, y1, 'r-')
plt.plot(x, y2, 'g--')
plt.plot(x, y3, 'b:')
plt.plot(x, np.cos(x)**2, 'y-.')

plt.title("Gráfico 4a")
plt.xlabel('x')
plt.ylabel("y")
plt.show()

# Modo 2:
# É possível alterar o estilo de cada linha individualmente
plt.plot(x,y1,'c-', x,y2,'m--', x,y3,'k:', x,np.cos(x)**2,'b-.')

plt.title("Gráfico 4b")
plt.xlabel('x')
plt.ylabel("y")
plt.show()
```

### Plotando pontos (símbolos)

* Ponto          = .
* Círculo        = o
* Quadrado       = s (*square*)
* Triângulo      = v ^ < > (cada um é uma rotação de triângulo)
* Quadrado       = s (*square*)
* Asterisco      = *
* Cruz           = +
* Xis            = x
* Traço baixo    = _
* Barra vertical = |

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando os dados:
x = np.arange(-np.pi, np.pi, 0.5)
y1 = x
y2 = np.sin(x)
y3 = np.cos(x)

# Modo 1:
# lw = nume width (lw=1, por default)
plt.plot(x, y1, 'ro')
plt.plot(x, y2, 'g.')
plt.plot(x, y3, 'bs')
plt.plot(x, np.cos(x)**2, 'y*')

plt.title("Gráfico 5a")
plt.xlabel('x')
plt.ylabel("y")
plt.show()

# Modo 2:
# É possível alterar o estilo de cada linha individualmente
plt.plot(x,y1,'c^', x,y2,'mv', x,y3,'k<', x,np.cos(x)**2,'b>')

plt.title("Gráfico 5b")
plt.xlabel('x')
plt.ylabel("y")
plt.show()
```

### Grades

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando os dados:
x = np.arange(-np.pi, np.pi, 0.5)
y1 = x
y2 = np.sin(x)
y3 = np.cos(x)

# Modo 1:
plt.plot(x, y1, 'r-o')
plt.plot(x, y2, 'g:.')
plt.plot(x, y3, 'bs-.')
plt.plot(x, np.cos(x)**2, 'y*--')

plt.grid(True)
plt.title("Gráfico 7a")
plt.xlabel('x')
plt.ylabel("y")
plt.show()

# Modo 2:
plt.plot(x,y1,'c^-', x,y2,'mv:', x,y3,'k<--', x,np.cos(x)**2,'b>-.')

plt.grid(True)  # Grade
plt.title("Gráfico 7b")
plt.xlabel('x')
plt.ylabel("y")
plt.show()
```

### Legendas

* Usa argumento da função **plt.plot()** label = "legenda".

* Para que apareçam, é preciso utilizar o comando **plt.legend()**

* Não funciona para o método que plota todas as curvas na mesma linha (modo 2).

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando os dados:
x = np.arange(-np.pi, np.pi, 0.5)
y1 = x
y2 = np.sin(x)
y3 = np.cos(x)

# Modo 1:
plt.plot(x, y1, 'r-o', label='f(x)=x')
plt.plot(x, y2, 'g:.', label='$f(x)=\sin(x)$')
plt.plot(x, y3, 'bs-.', label='$f(x)=\cos(x)$')
plt.plot(x, np.cos(x)**2, 'y*--', label='$f(x)=\cos(x)^2$')

plt.legend()    # Mostrar legendas
           
plt.grid(True)  # Grade
plt.title("Gráfico 8")
plt.xlabel('x')
plt.ylabel("y")
plt.show()
```

## Diagramas de dispersão

* Quando não existe à princípio uma relação funcional entre os dados.

Gerar uma distribuição aleatória de pontos distribuídos dentro de uma área circular com centro em (0,0) e raio igual a $r=1$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando os dados:
n = 1000
x = 2 * np.random.rand(n) - 1 # Gera números pseudo-aleatórios dentro de [-1,1)
y = 2 * np.random.rand(n) - 1

filtro = np.sqrt( x**2 + y**2 ) <= 1

x = x[filtro]
y = y[filtro]

# Visualizando:
# # Modo 1: funciona, mas o Python tem uma função específica
# plt.plot(x, y, '.')
# plt.show()

# Modo 2: plt.scatter()
plt.scatter(x, y, c='green', marker='.')  # 'g' ou 'green'
plt.title("Diagrama de Dispersão")
plt.ylabel('y')
plt.xlabel('x')

ax = plt.gca()
ax.set_aspect(1)  # Ajusta o aspectro: define a proporcionalidade entre os eixos y/x

plt.show()
```

## Gráfico de pizza

* ou **gráfico de setores** ou **gráfico circular**.

* Representa o quanto cada categoria contribui com o todo.

```python
import matplotlib.pyplot as plt

# Gerando dados:
tipos = ['Tipo 1', 'Tipo 2', 'Tipo 3', 'Tipo 4', 'Tipo 5']
valores = [10, 15, 27, 13, 5]

# Visualizando:
plt.pie(valores,
        labels = tipos,
        autopct = '%4.1f%%', # *Nota
        shadow = True)

# *O primeiro % indica que o que vem a seguir é especificação de formato, que é:
# o número total tem 4 dígitos, sendo 1 deles o ponto e 1 dígito após o ponto .
# f indica que é um float e o %% é para colocar o símbolo de porcentagem % 

plt.title("Diagrama de Pizza")
plt.show()
```

## Histogramas

* Variáveis contínuas (diferente do gráfico de barras que é para variáveis categóricas).

* Usa histograma quando a variável puder assumir qualquer valor dentro de um intervalo (não é discreto).

* No histograma as barras, chamadas de *bins*, são **coladas** uma na outra e não separadas como acontece em diagramas de barras.

* O critério para escolher o número de *bins* é utilizar no máximo raiz da quantidade de pontos. 

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando dados:
distr = np.random.randn(10000)  # randn o n significa que a distribuição é normal/gaussiana

# Visualizando:
# Histograma 1
plt.hist(distr)

plt.title("Histograma 1")
plt.xlabel("x")
plt.ylabel("Número de Ocorrências")
plt.show()

# Histograma 2
plt.hist(distr, density=True)

plt.title("Histograma 2")
plt.xlabel("x")
plt.ylabel("Densidade / Fração de Ocorrência")
plt.show()

# Histograma 3
plt.hist(distr, bins=50, color='yellow', density=True)  # Número de bins default é 10
# O critério para escolher o número de bins é utilizar no máximo raiz de n

plt.grid(True)
plt.title("Histograma 3")
plt.xlabel("x")
plt.ylabel("Densidade / Fração de Ocorrência")
plt.show()

# Histograma 4
plt.hist(distr, bins=50, color='red', density=True, histtype='step')

plt.title("Histograma 4")
plt.xlabel("x")
plt.ylabel("Densidade / Fração de Ocorrência")
plt.show()
```

## Múltiplos gráficos (em uma figura)



```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando dados:
distr = np.random.randn(10000)  # randn o n significa que a distribuição é normal/gaussiana

# Criando uma figura:
plt.figure()

# Visualizando:
# Histograma 1
plt.subplot(221)  # O arranjo total é 2x2 e estamos plotando o primeiro arranjo
plt.hist(distr)

plt.title("Histograma 1")
plt.xlabel("x")
plt.ylabel("Número de Ocorrências")

# Histograma 2
plt.subplot(222)
plt.hist(distr, density=True)

plt.title("Histograma 2")
plt.xlabel("x")
plt.ylabel("Densidade")

# Histograma 3
plt.subplot(223)
plt.hist(distr, bins=50, color='yellow', density=True)  # Número de bins default é 10
# O critério para escolher o número de bins é utilizar no máximo raiz de n

plt.grid(True)
plt.title("Histograma 3")
plt.xlabel("x")
plt.ylabel("Densidade") 

# Histograma 4
plt.subplot(224)
plt.hist(distr, bins=50, color='red', density=True, histtype='step')

plt.title("Histograma 4")
plt.xlabel("x")
plt.ylabel("Densidade")

# Ajustando os plots na figura:
plt.subplots_adjust(top = 0.92,
                    bottom = 0.08,
                    left = 0.10,
                    hspace = 0.55,
                    wspace = 0.5)

# Mostrando a figura:
plt.show()
```

## Texto em gráficos

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando dados:
mu = 3.5
sigma = 1.7
np.random.seed(0)
distr = sigma* np.random.randn(10000) + mu

plt.title("Distribuição Normal ($\mu, \, \sigma$)", fontsize=20, color='green')
plt.xlabel('x', fontsize=16, color='blue')
plt.ylabel("Densidade", fontsize=16, color='red')

plt.grid(True)

plt.text( -3,0.22, "Distribuição normal", fontsize=14)
plt.text( -3,0.18, "$\mu=3.5,\, \sigma=1.7$", fontsize=14)

# Visualização:
plt.hist(distr, bins=50, color='cyan', density=True)

plt.show()
```

```python
# Anotações em gráficos:
import matplotlib.pyplot as plt
import numpy as np

# Gerando dados: movimento oscilatório
A = 1.                     # Amplitude
f = 0.01                   # Frequência (Hz)
phi = 0.                   # Fase (rad)
t = np.arange(0, 1000, 1)  # Tempo em segundos

y = A * np.sin( 2 * np.pi * f * t + phi)

amort = np.exp(-0.005 * t)
y1 = amort * y

plt.figure()

# Gráfico 1: oscilação normal
plt.subplot(211)
plt.plot(t, y)
plt.title("Oscilação não amortecida")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude (m)")

# Gráfico 2: oscilação amortecida
plt.subplot(212)
plt.plot(t, y1)
plt.title("Oscilação amortecida")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude (m)")

plt.plot(t, amort, color='red')
plt.plot(t, -amort, color='red')

# Anotação
plt.annotate("Amortecimento",
            xy = (300,0.2),
            xytext=(650,0.7),
            arrowprops=dict(facecolor='black', shrink=0.15) 
            )

# Ajustando os plots na figura:
plt.subplots_adjust(top = 0.92,
                    bottom = 0.08,
                    left = 0.10,
                    hspace = 0.6,
                    wspace = 0.0)

plt.show()
```

## Gráficos logarítmicos

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando dados:
y = np.random.normal(loc=0.5, scale=0.4, size=1000)  # Permite entrar a média e a escala
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange( len(y) )
# plt.hist(y, bins=20)

plt.figure()

# Gráfico 1: linear-linear
plt.subplot(221)
plt.plot(x, y)
plt.xlabel("x (linear)")
plt.ylabel("y (linear)")
plt.title("Linear - Linear")

# Gráfico 2: linear-log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.xlabel("x (linear)")
plt.ylabel("y (log)")
plt.title("Linear - Log")

# Gráfico 3: log-log
plt.subplot(223)
plt.plot(x, y)
plt.yscale('log')
plt.xscale('log')
plt.xlabel("x (log)")
plt.ylabel("y (log)")
plt.title("Log - Log")

# Gráfico 4: log-linear
plt.subplot(224)
plt.plot(x, y)
plt.yscale('linear')
plt.xscale('log')
plt.xlabel("x (log)")
plt.ylabel("y (linear)")
plt.title("Linear - Log")

# Ajustando os plots na figura:
plt.subplots_adjust(top = 0.92,
                    bottom = 0.08,
                    left = 0.10,
                    hspace = 0.6,
                    wspace = 0.5  )

plt.show()
```

## Coordenadas polares

```python
import numpy as np
import matplotlib.pyplot as plt


r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```
