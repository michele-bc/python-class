# Classes

* As coisas em Python são objetos: listas, tuplas, funções, gráficos, etc. E cada objeto é uma instância, um exemplar, de uma determinada classe.

* Classe é uma estrutura que define quais os atributos e quais os métodos que os objetos contidos nela vão obter.

* Algumas classes que vimos:
  * int
  * float
  * str
  * list
  * dict
  * set

* É possível definir as próprias classes.

# Problema

Dado um ponto P com coordenadas cartesianas (x,y), desenvolva um código que:

* Identifique o quadrante do ponto
* Calcule suas coordenadas polares (r, $\theta$)
* Gere um gráfico mostrando sua posição


# Programação estruturada

* O algoritmo segue um fluxograma: o código é sequencial.
* Mas esse fluxo pode ser alterado com
  * estruturas de seleção: **if-elif-else** ou **try** e
  * estruturas iterativas: **for** ou **while**.

* Adequada para códigos relativamente pequenos (dezenas ou centenas de linhas).



```python
import math
import matplotlib.pyplot as plt

# Definir o ponto P:
P = (1, 1)  #@param

# Extrair as coordenadas x e y:
x = P[0]
y = P[1]

# Identificar o quadrante:
if x == 0 and y == 0:
  quadrante = 0
elif x >= 0 and y >= 0:
  quadrante = 1
elif x <= 0 and y >= 0:
  quadrante = 2
elif x <= 0 and y <= 0:
  quadrante = 3
else:
  quadrante = 4

# Calcular as coordenadas r e theta:
r = math.sqrt(x**2 + y**2)
theta = math.degrees( math.atan2( y, x ) )

# Gerar um gráfico:
plt.plot(x, y, 'o')

plt.xlim(-2,2)
plt.ylim(-2,2)

plt.plot( (-2,2), (0,0), 'k-')
plt.plot( (0,0), (-2,2), 'k-')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Point Position")

plt.text( 1.8,  1.8, '(1)')
plt.text(-1.8,  1.8, '(2)')
plt.text(-1.8, -1.8, '(3)')
plt.text( 1.8, -1.8, '(4)')

plt.text( x*1.1, y*1.1, "P")

plt.plot( (0,0.97*x), (0,0.97*y), 'r-' )

plt.show()

# Output:
print( "P =", P )
print( "x =", x )
print( "y =", y )
print( "O ponto P está no quadrante:", quadrante )
print( "r =", r )
print( "\u03B8 =", theta, 'deg' )
```

## Programação modular

* Programação funcional.

* Utiliza programação estruturada, mas ao invés de fazer um código inteiro com estrutura monolítica (*single-tiered*), o código é subdividido em peças menores, chamadas de **módulos** na teoria de programação (o sentido é diferente da termologia do Python). Esses módulos podem ser chamados por outros códigos e ser reutilizados.

* No Python esses módulos são as funções.

* Em C e R também. No Fortran são as subrotinas ou funções.

* É adequada quando o código tem tamanho de médio a grande, sendo mais fácil lidar com o código se ele estiver subdividido em peças menores e o programa principal fica reduzido.

* Outra situação em que é a programação modular é adequada é quando alguma tarefa é executada repetidamente. Assim, ao invés de escrevê-la várias vezes, escreve-se uma função que a executa e pode ser chamada diversas vezes.

```python
# Functions:

def identificaQuadrante(P):
  '''Identificar o quadrante do ponto P'''
  # Extrair as coordenadas x e y:
  x = P[0]
  y = P[1]

  if x == 0 and y == 0:
    quadrante = 0
  elif x >= 0 and y >= 0:
    quadrante = 1
  elif x <= 0 and y >= 0:
    quadrante = 2
  elif x <= 0 and y <= 0:
    quadrante = 3
  else:
    quadrante = 4
  return quadrante

def coordenadasPolares(P, unit='deg'):
  '''Calcular as coordenadas r e theta (em rad ou deg)'''
  import math
  import sys
  x = P[0]
  y = P[1]

  r = math.sqrt(x**2 + y**2)
  if unit == 'deg':
    theta = math.degrees( math.atan2( y, x ) )
  elif unit == 'rad':
    theta = math.atan2( y, x )
  else:
    print("ERROR: coordenadasPolares(): valor incorreto para unit.")
    sys.exit()
  return (r, theta)

def gerarGrafico(P):
  ''' Gerar um gráfico '''
  import matplotlib.pyplot as plt

  x = P[0]
  y = P[1]

  plt.plot(x, y, 'o')

  plt.xlim(-2,2)
  plt.ylim(-2,2)

  plt.plot( (-2,2), (0,0), 'k-')
  plt.plot( (0,0), (-2,2), 'k-')

  plt.xlabel("x")
  plt.ylabel("y")
  plt.title("Point Position")

  plt.text( 1.8,  1.8, '(1)')
  plt.text(-1.8,  1.8, '(2)')
  plt.text(-1.8, -1.8, '(3)')
  plt.text( 1.8, -1.8, '(4)')

  plt.text( x*1.1, y*1.1, "P")

  plt.plot( (0,0.97*x), (0,0.97*y), 'r-' )

  plt.show()

# -------------------------------------
# Main programe:
# -------------------------------------
P = (1, 1)  #@param
quadrante = identificaQuadrante(P)
polar = coordenadasPolares(P, unit='deg')
gerarGrafico(P)

# Output:
print( "P =", P )
print( "O ponto P está no quadrante:", quadrante )
print( "Coordenadas polares:", polar )
```

# Programação orientada a objetos

* Na programação estruturada o objetivo é modelar o processo. Na programação orientada a objetos, o foco é modelar o objeto.

* Na programação orientada a objetos, tenta-se identificar os objetos centrais do problema e definir quais devem ser seus atributos e métodos para que possam posteriormente resolver o problema.

* A vantagem é criar uma classe que contém todos os métodos (as funções da programação modular, p.e.) imbutidos nela.

* É possível então salvar essa classe, que é um módulo do Python, e qualquer problema do mesmo tipo pode ser resolvido com essa classe.

* É útil usar programação orientada a objetos quando você está trabalhando com objetos que têm atributos e métodos muito bem definidos.

### Métodos especiais

* Quando utilizamos **dir(objeto)** são listados os membros desse objeto. Os membros começam e terminam por '__' (dois traços baixos) são chamados de métodos especiais.

* Os métodos especiais são incorporados automaticamente quando classes são criadas.

* **dir()**: lista todos os membros (atributos e métodos) de um objeto.

* __new__ método construtor, que constrói o objeto quando a classe, p.e., **Point()** é chamada.
* __init__ método que realiza todas as inicializações necessárias após o objeto ser construído. P.e. atribui os valores *default* para x e y.
